### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
import random

def target_likelihood_analysis(df, target_col='fraud_reported'):
    # Filter categorical columns
    cat_cols = df.select_dtypes(include='object').columns.tolist()
    if target_col in cat_cols:
        cat_cols.remove(target_col)

    # Convert target to binary if it's 'Y'/'N'
    if df[target_col].dtype == 'object':
        df[target_col] = df[target_col].map({'Y': 1, 'N': 0})

    # Set plotting style
    sns.set(style="whitegrid")

    # Use a different color for each plot from a predefined palette
    palette_list = sns.color_palette("Set2") + sns.color_palette("pastel") + sns.color_palette("muted")

    for col in cat_cols:
        # Calculate fraud likelihood by category
        likelihood_df = df.groupby(col)[target_col].mean().reset_index()
        likelihood_df = likelihood_df.sort_values(by=target_col, ascending=False)

        # Pick a random color for the plot
        bar_color = random.choice(palette_list)

        # Plot
        plt.figure(figsize=(10, 4))
        sns.barplot(x=col, y=target_col, data=likelihood_df, color=bar_color)
        plt.title(f"Fraud Likelihood by {col}")
        plt.ylabel("Likelihood of Fraud (P(Y=1))")
        plt.xlabel(col)
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
```

```python
target_likelihood_analysis(train, target_col='fraud_reported')
```

### **6.1 Perform resampling** <font color="red">[3 Marks]</font>

Handle class imbalance in the training data by applying resampling technique.

Use the **RandomOverSampler** technique to balance the data and handle class imbalance. This method increases the number of samples in the minority class by randomly duplicating them, creating synthetic data points with similar characteristics. This helps prevent the model from being biased toward the majority class and improves its ability to predict the minority class more accurately.

**Note:** You can try other resampling techniques to handle class imbalance

### **6.2 Feature Creation** <font color="red">[4 marks]</font>

Create new features from existing ones to enhance the model's ability to capture patterns in the data. This may involve deriving features from date/time columns, combining features, or creating interaction terms.

```python
# Create new features based on your understanding for both training and validation data
def create_features(df):
    # Age groups
    df['age_group'] = pd.cut(df['age'], bins=[18, 30, 45, 60, 100], labels=['18-30', '31-45', '46-60', '60+'])

    # Claim ratios
    df['injury_to_total_ratio'] = df['injury_claim'] / (df['total_claim_amount'] + 1)
    df['vehicle_to_total_ratio'] = df['vehicle_claim'] / (df['total_claim_amount'] + 1)
    df['property_to_total_ratio'] = df['property_claim'] / (df['total_claim_amount'] + 1)

    # Sum of all claims
    df['claim_sum'] = df['injury_claim'] + df['vehicle_claim'] + df['property_claim']

    # High deductible flag
    df['high_deductible'] = (df['policy_deductable'] > 1000).astype(int)

    # Number of people involved (witnesses + bodily injuries + vehicles involved)
    df['people_involved'] = df['witnesses'] + df['bodily_injuries'] + df['number_of_vehicles_involved']

    return df

# Apply to train and val sets
train = create_features(train)
val = create_features(val)
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
def combine_categories(df):
    # Group hobbies: combine rare categories into 'other'
    hobby_counts = df['insured_hobbies'].value_counts()
    rare_hobbies = hobby_counts[hobby_counts < 30].index
    df['insured_hobbies'] = df['insured_hobbies'].replace(rare_hobbies, 'other')

    # Group occupations: combine rare categories
    occ_counts = df['insured_occupation'].value_counts()
    rare_occs = occ_counts[occ_counts < 30].index
    df['insured_occupation'] = df['insured_occupation'].replace(rare_occs, 'other')

    # Group auto_make: simplify common vs rare car brands
    df['auto_make'] = df['auto_make'].replace({
        'Mercedes': 'Luxury',
        'BMW': 'Luxury',
        'Chevrolet': 'Common',
        'Ford': 'Common',
        'Toyota': 'Common',
        'Dodge': 'Common',
        'Honda': 'Common',
        'Accura': 'Common',
        'Nissan': 'Common',
        'Jeep': 'Low-Risk',
        'Volkswagen': 'Low-Risk',
        'Saab': 'Low-Risk',
        'Suburu': 'Low-Risk',
        'Audi': 'Luxury',
        'Mazda': 'Low-Risk'
    })

    # Handle missing or ambiguous values
    df['authorities_contacted'] = df['authorities_contacted'].replace({'?': 'Unknown'})
    df['collision_type'] = df['collision_type'].replace({'?': 'Unknown'})

    return df

# Apply to both datasets
train = combine_categories(train)
val = combine_categories(val)
```

### **7.3 Find the Optimal Cutoff** <font color = red>[12 marks]</font>

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Define ROC function
def plot_roc_curve(y_true, y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    auc_score = roc_auc_score(y_true, y_pred_prob)

    plt.figure(figsize=(8, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()
```

```python
# Call the ROC function
plot_roc_curve(y_train, train_preds_df["predicted_prob"])
```

**Sensitivity and Specificity tradeoff**

After analysing the area under the curve of the ROC, check the sensitivity and specificity tradeoff to find the optimal cutoff point.

```python
# Define function to calculate metrics for multiple cutoff values
def evaluate_cutoffs(y_true, y_prob):
    cutoffs = np.arange(0.01, 1.0, 0.01)
    sensitivity = []
    specificity = []
    accuracy = []

    for cutoff in cutoffs:
        predicted_class = (y_prob >= cutoff).astype(int)
        tn, fp, fn, tp = metrics.confusion_matrix(y_true, predicted_class).ravel()

        sensitivity.append(tp / (tp + fn))
        specificity.append(tn / (tn + fp))
        accuracy.append((tp + tn) / (tp + tn + fp + fn))

    return pd.DataFrame({'cutoff': cutoffs,'sensitivity': sensitivity,'specificity': specificity,'accuracy': accuracy})

# Get metrics across cutoffs
cutoff_df = evaluate_cutoffs(y_train, train_preds_df['predicted_prob'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cutoff_df['cutoff'], cutoff_df['sensitivity'], label='Sensitivity', color='green')
plt.plot(cutoff_df['cutoff'], cutoff_df['specificity'], label='Specificity', color='blue')
plt.plot(cutoff_df['cutoff'], cutoff_df['accuracy'], label='Accuracy', color='red')
plt.xlabel('Cutoff Threshold')
plt.ylabel('Score')
plt.title('Sensitivity, Specificity, and Accuracy at Different Cutoffs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute the absolute difference between sensitivity and specificity
cutoff_df['diff'] = abs(cutoff_df['sensitivity'] - cutoff_df['specificity'])

# Find the row where the difference is smallest
optimal_cutoff_row = cutoff_df.loc[cutoff_df['diff'].idxmin()]
optimal_cutoff = optimal_cutoff_row['cutoff']
```

```python
print(f"Optimal Cutoff: {optimal_cutoff:.2f}")
print(f"Sensitivity: {optimal_cutoff_row['sensitivity']:.4f}")
print(f"Specificity: {optimal_cutoff_row['specificity']:.4f}")
print(f"Accuracy: {optimal_cutoff_row['accuracy']:.4f}")
```
